```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE
```

```python
x,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=2,weights=[0.90],random_state=12)
```

```python
df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['Target'])
df_final=pd.concat([df1,df2],axis=1)
```

```python
df_final['Target'].value_counts()
```

```python
plt.scatter(df_final['f1'],df_final['f2'],c=df_final['Target'])
plt.savefig('Before-SMOTE.png', dpi=300, bbox_inches='tight')
```

## Transforming Dataset

```python
oversample=SMOTE()
x,y=oversample.fit_resample(df_final[['f1','f2']],df_final['Target'])
```

```python
x.shape
```

```python
y.shape
```

```python
df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['Target'])
df_final=pd.concat([df1,df2],axis=1)
```

```python
plt.scatter(df_final['f1'],df_final['f2'],c=df_final['Target'])
plt.savefig('After-SMOTE.png', dpi=300, bbox_inches='tight')
```
